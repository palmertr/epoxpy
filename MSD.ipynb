```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("anneal-b1e5-long_0.01/msd.log")
x = data[:,0]
a = data[:,1]
b = data[:,2]
c = data[:,3]
plt.plot(x,a, label="A", color="red")
plt.plot(x,b, label="B", color="blue")
plt.plot(x,c, label="C", color="green")
plt.legend(loc='best', shadow=False, prop={'size':20})
plt.xlabel("Time Step")
plt.ylabel(r"MSD")
```

```python
data = np.genfromtxt("anneal-b1e5-long_0.01/out.log")
x = data[:,0]
temp = data[:,6]
plt.plot(x,temp, label="temp", color="red")
plt.xlabel("Time Step")
plt.ylabel("Temp")
```
