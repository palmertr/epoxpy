```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

msd_data = np.genfromtxt(msd_path,skip_header=True)
log_data = np.genfromtxt(log_path,skip_header=True)

fig,ax1 = plt.subplots(figsize=(10,7))
ax2 = ax1.twinx()
ax1.plot(msd_data[:,0],msd_data[:,1],linestyle='--',color='r',label='msd(Amine)',zorder=10)
ax1.plot(msd_data[:,0],msd_data[:,2],color='b',label='msd(Epoxy)',zorder=9)
#ax1.plot(msd_data[:,0],msd_data[:,3],color='g',label='msd(Toughner)',zorder=8)
ax2.plot(log_data[:,0],log_data[:,6],color='g',label='temperature',zorder=2)

ax1.set_xlabel('time',fontsize=15)
ax1.set_ylabel('msd',fontsize=15)

x1,x2,y1,y2 = ax1.axis()
#x1=0.7e7
ax1.axis((x1,x2,-y1,y2))
ax1.legend(fontsize=15,loc="center left")
ax2.legend(fontsize=15,loc="center right")
ax2.set_ylabel('temperature',fontsize=15)
```
