```python
%matplotlib inline
import matplotlib
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bond_test(kT, E):
    mb_stats = np.exp(-E/kT)
    #print("mb_stats",mb_stats)
    rand = random.random()
    #print("rand",rand)
    if mb_stats > rand:
        return 1
    else:
        return 0
def bond_test_rank(kT, E, bond_rank):
    mb_stats = np.exp(-E/kT)
    #print("mb_stats",mb_stats)
    rand = random.random()
    #print("rand",rand)
    if mb_stats/float(bond_rank) > rand:
        return 1
    else:
        return 0
    
```

```python
temps = [10, 1, 0.1, 0.01]
e_vals = [10, 1, 0.1, 0.01]
#temps = [0, 1, 2, 3, 4, 5]
temps_per = []
for e in e_vals:
    for t in temps:
        bond_percent = []
        for i in range(int(1e6)):
            bond_percent.append(bond_test(t, e))
        temps_per.append(np.mean(bond_percent))
plt.xscale('log')
plt.xlabel("kT")
plt.ylabel("Probability")
plt.plot(temps, temps_per[0:4], label=e_vals[0], ls="--", marker="x")
plt.plot(temps, temps_per[4:8], label=e_vals[1], ls="--", marker="x")
plt.plot(temps, temps_per[8:12], label=e_vals[2], ls="--", marker="x")
plt.plot(temps, temps_per[12:16], label=e_vals[3], ls="--", marker="x")
plt.legend(loc="best")
```
