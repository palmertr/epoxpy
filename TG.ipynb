```python
%matplotlib inline
import re
import numpy as np
import matplotlib.pyplot as plt

def get_pe(fname, cut):
    data = np.genfromtxt(fname)
    pe = data[:,1][cut:]
    x = data[:,0][cut:]
    n = len(pe)
    mean = np.mean(pe)
    std = np.std(pe)
    err = std/np.sqrt(n)
    return mean, err, x

def gen_plot(tupes):
    fig, ax = plt.subplots()
    regex = re.compile('\d\.\d+')
    for plots in tupes:
        files, color, label = plots
        temp_list = []
        pe_list = []
        err_list = []
    
        for f in files:
            pe, err, x = get_pe(f,101)
            pe_list.append(pe)
            err_list.append(err)
            temp_list.append(regex.findall(f)[0])
            
        ax.errorbar(temp_list, pe_list, yerr=err_list, color=color, 
                    marker = 'o', markeredgewidth=0, markersize=4, 
                    ls='dashed', linewidth=1, label=label)
    
    #fig.rcParams.update({'mathtext.default':  'regular' })
    ax.set_xlabel(r"Temperature (K)")
    ax.set_ylabel(r"PE")
    ax.yaxis.major.formatter._useMathText = True
    ax.legend(loc='best', shadow=False, prop={'size':20})

files1 = ["runs/kestrel/msd-bonding-1e5-kT-0.1/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-0.5/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-1.0/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-1.5/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-2.0/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-2.5/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-3.0/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-3.5/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-4.0/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-4.5/out.log",
         "runs/kestrel/msd-bonding-1e5-kT-5.0/out.log",
        ]
files2 = ["runs/kestrel/msd-bonding-2e5-kT-0.1/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-0.5/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-1.0/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-1.5/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-2.0/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-2.5/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-3.0/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-3.5/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-4.0/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-4.5/out.log",
         "runs/kestrel/msd-bonding-2e5-kT-5.0/out.log",
        ]

files3 = ["runs/kestrel/msd-no-bonding-kT-0.05/out.log",
         "runs/kestrel/msd-no-bonding-kT-0.1/out.log",
         "runs/kestrel/msd-no-bonding-kT-0.5/out.log",
         "runs/kestrel/msd-no-bonding-kT-1.0/out.log",
         "runs/kestrel/msd-no-bonding-kT-1.5/out.log",
         "runs/kestrel/msd-no-bonding-kT-2.0/out.log",
         "runs/kestrel/msd-no-bonding-kT-2.5/out.log",
         "runs/kestrel/msd-no-bonding-kT-3.0/out.log",
         "runs/kestrel/msd-no-bonding-kT-3.5/out.log",
         "runs/kestrel/msd-no-bonding-kT-4.0/out.log",
         "runs/kestrel/msd-no-bonding-kT-4.5/out.log",
         "runs/kestrel/msd-no-bonding-kT-5.0/out.log",
        ]

files4 = ["runs/kestrel/msd-bonding-3e5-kT-0.1/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-0.5/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-1.0/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-1.5/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-2.0/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-2.5/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-3.0/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-3.5/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-4.0/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-4.5/out.log",
         "runs/kestrel/msd-bonding-3e5-kT-5.0/out.log",
        ]



gen_plot([(files1, "g", "1e5"),(files2, "r", "2e5"),(files3, "y", "No Bonding"),(files4, "b", "3e5")])
```
